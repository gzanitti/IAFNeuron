SimulaciÃ³n de neurona de Izhikevich con inputs de 100 sinapsis con disparos Poisson.

```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 
%pylab inline
```

```python
# 1) Inicializar parametros
tmax = 1000
dt = 0.5
```

```python
# 1.1) Parametros de red / neurona
a = 0.02 #RS,IB: 0.02, FS: 0.1
b = 0.2 #RS, IB, FS: 0.2
c = -65 #RS, FS: -65 IB: -55
d = 8. #RS: 8, IB: 4, FS: 2
tau_s = 10 #Decaimiento de la sinapsis [ms]
tau_d = 500 #Depresion sinaptica [ms]
stp_u = 0.5 #Parametro STP
```

```python
# 1.2) Parametros de entrada
tr = array([200.,700])/dt
rate_in = 10 #Tasa de disparo de entrada
n_in = 100 #Cantidad de inputs
w_in = 0.03 #Peso de los inputs
W_in = w_in * ones(n_in) #Vector
```

```python
# 2) Reservar memoria
T = ceil(tmax/dt)
v = zeros(T)
u = zeros(T)
v[0] = -70 #Potencial de reposo
u[0] = -14 #Estado estable
s_in = zeros(n_in)
E_in = zeros(n_in)
prate = dt * rate_in * 1e-3
h = ones(n_in)
lastsp = -infty * ones(n_in)
```

```python
# 3) Iteracion sobre el tiempo
for t in arange(T-1):
    # 3.1) Leo la entrada
    if t > tr[0] and t < tr[1]:
        p = uniform(size=n_in)<prate
        
        #Actualizar depresion sinaptica
        tmp = exp(dt * (lastsp[p] - t)/tau_d)
        h[p] = 1 - (1 + (stp_u - 1) * h[p]) * tmp
        lastsp[p] = t
    else:
        p = 0
        
    #Calcular la corriente de entrada
    s_in = (1 - dt/tau_s) * s_in + p
    I = dot(W_in, s_in * E_in)
    I -= dot(W_in, s_in) * v[t]
    
    if v[t] < 35:
        # 3.2 Actualizar ODE
        dv = (0.04 * v[t] + 5) * v[t] + 140 - u[t]
        v[t+1] = v[t] + (dv + I) * dt
        du = a * (b * v[t] - u[t])
        u[t+1] = u[t] + dt * du
    else:
        # 3.3 Spike!
        v[t] = 35
        v[t+1] = c
        u[t+1] = u[t] + d        
```

```python
# 4) Ploteo los resultados
figure()
tvec = arange(0., tmax, dt)
plot(tvec, v, 'b', label='Voltage')
xlabel('Tiempo [ms]')
ylabel('Voltage de la membrana [mV]')
title("Neurona simple con %d inputs Poisson y STP" % n_in)
show()
```
